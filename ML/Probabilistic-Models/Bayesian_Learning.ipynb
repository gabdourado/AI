## **Aprendizado Bayesiano**

Em **classificação**, o Aprendizado Bayesiano consiste em atualizar a probabilidade de uma classe $y$ após observar um dado $\mathbf{x}$.


### **Teorema de Bayes**

O núcleo desse processo é dado por:

$$
P(y \mid \mathbf{x}) = \dfrac{P(\mathbf{x} \mid y)\,P(y)}{P(\mathbf{x})}
$$

#### **Posterior**: $P(y \mid \mathbb{x})$

É a probabilidade da classe $y$ dado o dado observado $\mathbf{x}$:

$$
y \mid \mathbf{x} \sim P(y \mid \mathbf{x})
$$

Representa a crença atualizada sobre a classe.

#### **Likelihood (Verossimilhança)**: $P(\mathbf{x} \mid y)$

É a probabilidade de observar o dado $\mathbf{x}$ assumindo que a classe é $y$:

$$
\mathcal{L}(y) = P(\mathbf{x} \mid y)
$$

Mede o quão bem a classe $y$ explica o dado observado.

#### **Prior**: $P(y)$

É a probabilidade da classe antes de observar os dados:

$$
y \sim P(y)
$$

Representa a frequência ou crença inicial sobre cada classe.

#### **Evidence (Marginal Likelihood)**: $P(\mathbf{x})$

É um fator de normalização, dado por:

$$
P(\mathbf{x}) = \sum_{y} P(\mathbf{x} \mid y)\,P(y)
$$

Garante que $P(y \mid \mathbf{x})$ seja uma distribuição válida.

#### **Forma utilizada na prática**

Como $P(\mathbf{x})$ é constante para todas as classes, utilizamos:

$$
P(y \mid \mathbf{x}) \propto P(\mathbf{x} \mid y)\,P(y)
$$

E a decisão é feita por:

$$
\hat{y} = \arg\max_y P(\mathbf{x} \mid y)\,P(y)
$$

Esse critério de decisão é conhecido como **Máximo a Posteriori (MAP)**.

### **Gaussian Naive Bayes**

#### **Fundamentação**

Neste modelo, assumimos que, para cada classe $c$ e para cada feature $x_i$, temos uma Gaussiana com **média** e **variância** próprias:

$$
x_i \mid y = c \sim \mathcal{N}(\mu_{c,i}, \sigma_{c, i}^{2})
$$

#### **Hipótese de Naive Bayes**

Assumimos que as features são independentes dado a classe:

$$
P(x_1, x_2, x_3, \dots, x_n \mid y) = \prod_{i=1}^{n}P(x_i \mid y)
$$

#### **Modelo Probabilístico**

Modelamos então probabilidade de uma amostra $x = (x_1, x_2, x_3, \dots, x_n)$ ser da classe $c$ como:

$$
P(y = c \mid x) \propto P(y = c)\prod_{i=1}^{n}P(x_i \mid y = c)
$$

onde:

*   $P(y = c)$ é o **prior**
*   $P(x_i \mid y = c)$ é o **likehood** e é dado pela distribuição normal:

$$
P(x_i \mid y = c) = \dfrac{1}{\sqrt{2\pi\sigma_{c,i}^{2}}} \exp \left(-\dfrac{(x_i - \mu_{c,i})}{2\sigma_{c,i}^{2}}\right)
$$

#### **Decisão da classe**

A predição é feita escolhendo a classe que maximiza a probabilidade posterior:

$$
\hat{y} = \arg\max_c P(y = c)\prod_{i=1}^{n}P(x_i \mid y = c)
$$

#### **Aplicação do logaritmo**

Como as probabilidades condicionais são tipicamente muito pequenas,

$$
P(x_i \mid y) \ll 1
$$

o produto dessas probabilidades pode rapidamente tender a zero:

$$
\prod_{i=1}^{n} P(x_i \mid y) \rightarrow 0
$$

causando problemas de underflow numérico.

Para contornar isso, aplicamos o logaritmo, que transforma produtos em somas e melhora a estabilidade numérica:

$$
\hat{y} = \arg\max_c \left[ \log P(y = c) + \sum_{i=1}^{n} \log P(x_i \mid y = c) \right].
$$

Como o logaritmo é uma função monotonicamente crescente, ele não altera o resultado do $\arg\max$.
