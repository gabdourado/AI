## **Aprendizado Bayesiano**

O Aprendizado Bayesiano em ML é em um método que atualiza probabilidades de hipóteses à medida que novos dados são observados.

### **Teorema de Bayes**

O núcleo do aprendizado bayesiano é o seguinte teorema:

$$
P(\theta | D) = \dfrac{P(D | \theta)P(\theta)}{P(D)}
$$

#### **Posterior**: $P(\theta | D)$

É a distribuição de probabilidade atribuída aos parâmetros após a observação de novos dados:

$$
\theta | D \sim P(\theta)
$$

#### **Likehood (Verissimilhança)**: $P(D | \theta)$

É a função de probabilidade dos dados em função dos parâmetros:

$$
L(\theta) = P(D | \theta)
$$

#### **Prior**: $P(\theta)$

É a distribuição de probabilidade atribuída aos parâmetros antes a observação de novos dados:

$$
\theta \sim P(\theta)
$$

#### **Evidence**: $P(D)$

É um fator de normalização, chamado também de marginal likehood, sendo dado por:

$$
P(D) = \int P(D | \theta)P(\theta)d\theta
$$

### **Forma não formalizada**

Diz-se costumeiramente que:

$$
P(\theta | D) \propto P(D | \theta)P(\theta)
$$

pois o denominador, $P(D)$, é apenas uma contante de normalização.
