## **Árvores de Decisão (Decision Trees)**

É um algoritmo de **aprendizado supervisionado**, usado para tarefas de classificação e regressão ao particionar recursivamente o espaço de características em regiões puras. Neste notebook, é abordada a aplicação do algoritmo para terefa de classificação.

Dado um conjunto de observações

$$
\mathcal{D} = \{(\mathbf{x}_i, y_i)\}_{i=1}^N
\quad \text{onde}
\quad \mathbf{x}_i \in \mathbb{R}^{D} \text{ e } y_i \in \{1, \ldots, K\}
$$

uma árvore de decisão $T$ é definida como um mapeamento

$$
T : \mathbb{R}^{D} \to \{1, \ldots, K\}
$$

construído por meio de um conjunto de **regiões disjuntas** $\{R_m\}_{m=1}^{M}$ que particionam $\mathbb{R}^D$, tal que

$$
\bigcup_{m=1}^{M} R_m = \mathbb{R}^{D}
\qquad \text{e} \qquad
R_m \cap R_{m'} = \emptyset \;\; \forall\, m \neq m'.
$$

A predição para uma nova observação $\mathbf{x}$ é a **classe majoritária** na região $R_m$ que contém $\mathbf{x}$:

$$
\hat{y} = \underset{k}{\arg\max} \;\; \hat{p}_{mk}
$$

onde $\hat{p}_{mk}$ é a proporção de exemplos da classe $k$ na região $R_m$.

A construção da árvore se apoia em dois conceitos fundamentais:

*   **Critério de impureza**: mede a heterogeneidade de um nó
*   **Critério de divisão**: seleciona o atributo e o ponto de corte que mais reduzem a impureza

### **Critérios de Impureza**

Seja $\hat{p}_{mk}$ a proporção de exemplos da classe $k$ no nó $m$. Dois critérios clássicos são:

#### **Entropia**

$$
H(\hat{p}_m) = -\sum_{k=1}^{K} \hat{p}_{mk} \log_2 \hat{p}_{mk}
$$

A entropia é zero quando o nó é puro (uma única classe) e máxima quando todas as classes são equiprováveis.

O **Ganho de Informação** de um atributo $j$ com ponto de corte $t$ após dividir um nó $m$ é:

$$
IG(j) = H(\hat{p}_m) - \left[\frac{|R_L|}{|R_m|}H(\hat{p}_L) + \frac{|R_R|}{|R_m|}H(\hat{p}_R)\right]
$$

onde $R_L$ e $R_R$ são os subconjuntos esquerdo e direito gerados pela divisão.

#### **Índice de Gini**

$$
\text{Gini}(\hat{p}_m) = \sum_{k=1}^{K} \hat{p}_{mk}(1 - \hat{p}_{mk}) = 1 - \sum_{k=1}^{K} \hat{p}_{mk}^{2}
$$
É a probabilidade de classificar incorretamente um exemplo escolhido ao acaso, caso o rótulo seja sorteado segundo a distribuição empírica $\hat{p}_m$.

### **Algoritmo de Construção**

O algoritmo constrói a árvore de forma **gulosa** e **recursiva**.

Em cada nó $m$ contendo o subconjunto $\mathcal{D}_m$, busca-se a divisão binária ótima:

$$
(j^*, t^*) = \underset{j,\, t}{\arg\min}
\left[
  \frac{|R_L(j,t)|}{|\mathcal{D}_m|}\, Q(R_L) +
  \frac{|R_R(j,t)|}{|\mathcal{D}_m|}\, Q(R_R)
\right]
$$

onde $Q(\cdot)$ é o critério de impureza escolhido (Gini ou Entropia), e

$$
R_L(j, t) = \{\mathbf{x} \in \mathcal{D}_m \mid x_j \leq t\}
\qquad
R_R(j, t) = \{\mathbf{x} \in \mathcal{D}_m \mid x_j > t\}.
$$

Para cada atributo $j \in \{1,\ldots,D\}$, os candidatos $t$ são todos os valores medianos entre **observações consecutivas ordenadas**.

Um ponto interessante é que **Facelli (2011)** maximiza $IG(j, t)$, o que  é equivalente a minimizar a impureza ponderada dos filhos, ou seja:

$$
(j^*, t^*) = \underset{j,\, t}{\arg\max} \,\, IG(j, t)= \underset{j,\, t}{\arg\min}
\left[
  \frac{|R_L(j,t)|}{|\mathcal{D}_m|}\, Q(R_L) +
  \frac{|R_R(j,t)|}{|\mathcal{D}_m|}\, Q(R_R)
\right]
$$

### **Critérios de Parada**

A recursão é interrompida quando ao menos uma das condições é satisfeita:

*   O nó é **puro**: $\max_k \hat{p}_{mk} = 1$
*   O número de exemplos no nó é menor que um limiar: $|\mathcal{D}_m| < n_{\min}$
*   A profundidade máxima $d_{\max}$ foi atingida
*   A redução de impureza após a melhor divisão é inferior a $\epsilon > 0$

Ao parar, o nó torna-se uma **folha** com rótulo $\hat{y}_m = \arg\max_k \hat{p}_{mk}$.
