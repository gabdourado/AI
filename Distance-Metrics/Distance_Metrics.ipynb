## Distance Metrics for Machine Learning (ML)

### **Definição**

Dado um conjunto $\mathcal{X}$ e $d: \mathcal{X} \times \mathcal{X} \rightarrow \mathbb{R}^{+}$ uma função que dados $x, y$ e $z \in \mathcal{X}$, satisfaz os seguintes axioma.

*   $d(x, y) = 0$, se e somente se, $x = y$
*   $d(x, y) = d(y, x)$ **(Simetria)**
*   $d(x, z) \leq d(x, y) + d(y, z)$ **(Desigualdade Triangular)**

Dizemos que $d$ é a métrica ou distância em $\mathcal{X}$.


### **Distância Minkowski**

Consiste em uma métrica de distância generalizada, dada pela fórmula:


$$
d(x,y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{\frac{1}{p}}
$$

observe que:
*   se $p = 1$, temos a **distância Manhattan (L1)**
*   se $p = 2$, temos a **distância Euclidiana (L2)**


### **Distância Cosine**

Essa métrica vem da similaridade do cosseno (cosine similarity), que mede o ângulo entre dois vetores:

$$
\text{cosine similarity}(x,y) = \dfrac{\langle x,y \rangle}{\|x\| \cdot \|y\|}
$$

Definimos a cosine distance como sendo:

$$
\text{cosine distance}(x,y) = 1 - \text{cosine similarity}(x,y)
$$

Note que essa função não é, em geral, uma métrica no sentido matemático estrito,
pois pode violar a desigualdade triangular.

### **Distância Mahalanobis**

Essa métrica considera a variância e a correlação entre os dados. Calculando a distância entre um vetor e uma distribuição:

$$
d(x, y) = \sqrt{(x - y)^{\top}\Sigma^{-1}(x - y)}
$$

onde $\Sigma$ é a matriz de covariância.

### **Distância Hamming**

Essa métrica mede quantas posições diferentes existem entre duas sequências de mesmo tamanho:

$$
d(x, y) = \#(x_i \neq y_i)
$$

sendo usada em dados binários ou categóricos.
