## **Introdução à Teoria das Probabilidades**

Essas são as anotações de conceitos base para o estudo das Probabilidades aplicado a Machine Learning.

### **Espaço Amostral e Evento**

Dado um experimento aleatório, define-se o **espaço amostral** $\Omega$ como o conjunto de todos os resultados possíveis.

Uma $\sigma$-álgebra $\mathcal{F}$ sobre $\Omega$ é uma coleção de subconjuntos de $\Omega$ tal que:

$$
\Omega \in \mathcal{F}, \quad A \in \mathcal{F} \Rightarrow A^{c} \in \mathcal{F}, \quad A_1, A_2, \dots \in \mathcal{F} \Rightarrow
\bigcup_{i=1}^{\infty} A_i \in \mathcal{F}
$$

Qualquer $A \in \mathcal{F}$ é chamado de **evento**. A tripla $(\Omega, \mathcal{F}, P)$ é chamado espaço de probabilidade.

### **Axiomas de Kolmogorov**

Uma medida de probabilidade é uma função $P: \mathcal{F} \to \mathbb{R}$ que satisfaz:

* **Não Negativa**

Para qualquer evento $A \in \mathcal{F}$, segue que:

$$
P(A) \geq 0
$$

* **Normalização**

A probabilidade do espaço amostral $\Omega$ é:

$$
P(\Omega) = 1
$$

* **$\sigma$-Aditividade**

Para toda sequência $(A_i)_{i \in \mathbb{N}}$ de eventos mutualmente exclusivos, ou seja:

$$
A_i \cap A_j = \emptyset \quad \forall i \neq j
$$

então segue que:

$$
P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i)
$$

### **Probabilidade Clássica**

Quando $\Omega$ é finito e todos os resultados são equiprováveis, para qualquer evento $A \in \mathcal{F}$:

$$
P(E) = \dfrac{|A|}{|\Omega|}
$$

### **Probabilidade Condicional**

Dados $A,B \in \mathcal{F}$ com $P(B) > 0$, a probabilidade condicional de $A$ dado $B$ é:

$$
P(A \mid B) = \dfrac{P(A \cap B)}{P(B)}
$$

### **Independência**

Dois eventos $A, B \in \mathcal{F}$ são ditos independentes se e somente se:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

Isso significa que a ocorrência de $A$ não influencia na ocorrencia de $B$.

### **Teorema da Probabilidade Total**

Seja $(B_i)_{i=1}^{n} \subset \mathcal{F}$ uma partição de $\Omega$, isto é:

$$
\bigcup_{i=1}^{n} B_i = \Omega \quad \text{e} \quad B_i \cap B_j = \emptyset \quad \forall i \neq j
$$

om $P(B_i) > 0$ para todo $i$. Então, para qualquer evento $A \in \mathcal{F}$:

$$
P(A) = \sum_{i=1}^{n} P(A \mid B_i)P(B_i)
$$
