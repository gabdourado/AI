### **Distribuições de Probabilidade**

Seguem anotações sobre distrições de Probabilidade, muito utilizadas em Machine Learning.

### **Variáveis Aleatórias**

Dado um espaço de probabilidade $(\Omega, \mathcal{F}, P)$, uma **variável aleatória** é uma função:

$$X : \Omega \to \mathbb{R}$$

que associa a cada resultado $\omega \in \Omega$ um valor numérico $X(\omega) \in \mathbb{R}$.

A **função de distribuição acumulada (CDA)** de $X$ é definida como:

$$F_X(x) = P(X \leq x), \quad \forall x \in \mathbb{R}$$

e caracteriza completamente a distribuição de $X$.

#### **Variáveis Discretas**

A variável $X$ é dita **discreta** se assumir valores em um conjunto enumerável $\mathcal{X} \subseteq \mathbb{R}$.

Sua distribuição é caracterizada pela **função de massa de probabilidade (PMF)**:

$$p_X(x) = P(X = x), \quad x \in \mathcal{X}$$

que satisfaz $p_X(x) \geq 0$ e $\displaystyle\sum_{x \in \mathcal{X}} p_X(x) = 1$.

#### **Variáveis Contínuas**

A variável $X$ é dita **contínua** se existir uma função $f_X : \mathbb{R} \to [0, \infty)$, chamada de **função densidade de probabilidade (PDF)**, tal que:

$$P(a \leq X \leq b) = \int_a^b f_X(x)\, dx$$

A PDF satisfaz $\displaystyle\int_{-\infty}^{\infty} f_X(x)\, dx = 1$. Note que $f_X(x)$ não é uma probabilidade, pois essa só é obtida integrando sobre um intervalo.

### **Esperança**

Consiste no valor esperado de uma variável aleatória:

* **Caso Discreto**

Sendo $X$ uma variável discreta, então:

$$
E[X] = \sum_{x \in \mathcal{X}} x\, p_X(x)
$$

* **Caso Contínuo**

Sendo $X$ uma variável contínua, então:

$$
E[X] = \displaystyle\int_{-\infty}^{\infty} x \, f_{X}(x)dx
$$

### **Variância**

Seja $\mu = E[X]$, a variância de $X$ mede a dispersão dos valores em torno da média:

$$
Var(X) = E[(X - \mu)^{2}]
$$

Uma forma computacional equivalente de determinar a variância é:

$$
Var(X) = E[X^{2}] - E[X]^{2}
$$

É comum denotar $\text{Var}(X) = \sigma_X^2$.


### **Desvio Padrão**

Como $Var(X)$ está em unidades ao quadrado de $X$, define-se o
desvio padrão como:

$$
\sigma_{X} = \sqrt{Var(X)}
$$

que possui a mesma unidade de $X$, tornando-o mais interpretável como medida de dispersão.


### **Distribuições de Probabilidade**

Uma **distribuição de probabilidade** especifica como a probabilidade é atribuída aos valores de uma variável aleatória $X$.

#### **Bernoulli**

Uma variável aleatória segue a distribuição de Bernouli se:

$$
X \in \{0, 1\}
$$

sendo denotada por:

$$
X \sim \text{Bernoulli}(p)
$$

Possui uma PMF dada por:

$$
p_X(x) = p^{x}(1 - p)^{1-x}
$$

Possui esperança e variância dadas por:

$$E[X] = p \qquad Var(X) = p(1 -p)$$

#### **Binomial**

Uma variável aleatória $X$ segue uma distribuição Binomial com parâmetros: $n \in \mathbb{N}$ e $p \in (0,1)$, denotada por:

$$
X \sim \text{Binomial}(n, p)
$$

se $X$ conta o número de sucessos em $n$ ensaios de Bernoulli independentes e identicamente distribuídos.

Sua PMF é:

$$p_X(x) = \binom{n}{x} p^x (1-p)^{n-x}, \quad x \in \{0, 1, \dots, n\}$$

onde $\dbinom{n}{x} = \dfrac{n!}{x!(n-x)!}$ é o coeficiente binomial.

Sua esperança e variância são:

$$E[X] = np \qquad \text{Var}(X) = np(1-p)$$

#### **Multinomial**

Uma variável aleatória vetorial

$$\mathbf{X} = (X_1, \dots, X_k)$$

segue uma distribuição Multinomial com parâmetros $n \in \mathbb{N}$ e $\mathbf{p} = (p_1, \dots, p_k)$ com $p_i > 0$ e $\sum_{i=1}^k p_i = 1$, denotada:

$$\mathbf{X} \sim \text{Multinomial}(n, \mathbf{p})$$

se $\mathbf{X}$ conta o número de ocorrências de cada uma das $k$ categorias em $n$ ensaios independentes.

Sua PMF conjunta é:

$$p_{\mathbf{X}}(x_1, \dots, x_k) = \frac{n!}{x_1! \cdots x_k!} \prod_{i=1}^{k} p_i^{x_i}, \quad x_i \in \{0, 1, \dots, n\},\quad \sum_{i=1}^k x_i = n$$

Para cada componente:

$$E[X_i] = np_i \qquad \text{Var}(X_i) = np_i(1 - p_i)$$

#### **Normal (Gaussiana)**

Uma variável aleatória $X$ segue uma distribuição Normal com parâmetros $\mu \in \mathbb{R}$ e $\sigma^2 > 0$, denotada por:

$$X \sim \mathcal{N}(\mu, \sigma^2)$$

se sua PDF é:

$$f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad x \in \mathbb{R}$$

Sua esperança e variância são:

$$E[X] = \mu \qquad \text{Var}(X) = \sigma^2$$

O caso $\mathcal{N}(0, 1)$ é chamado de **Normal Padrão**, frequentemente denotada $Z$, e sua CDA é:

$$\Phi(z) = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}} e^{-t^2/2} \, dt$$

Qualquer $X \sim \mathcal{N}(\mu, \sigma^2)$ pode ser padronizada como $Z = \dfrac{X - \mu}{\sigma} \sim \mathcal{N}(0,1)$.
