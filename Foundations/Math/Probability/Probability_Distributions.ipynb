### **Distribuições de Probabilidade**

Seguem anotações sobre distrições de Probabilidade, muito utilizadas em Machine Learning.

### **Variáveis Aleatórias**

Dado um espaço de probabilidade $(\Omega, \mathcal{F}, P)$, uma **variável aleatória** é uma função:

$$X : \Omega \to \mathbb{R}$$

que associa a cada resultado $\omega \in \Omega$ um valor numérico $X(\omega) \in \mathbb{R}$.

A **função de distribuição acumulada (CDA)** de $X$ é definida como:

$$F_X(x) = P(X \leq x), \quad \forall x \in \mathbb{R}$$

e caracteriza completamente a distribuição de $X$.

#### **Variáveis Discretas**

A variável $X$ é dita **discreta** se assumir valores em um conjunto enumerável $\mathcal{X} \subseteq \mathbb{R}$.

Sua distribuição é caracterizada pela **função de massa de probabilidade (PMF)**:

$$p_X(x) = P(X = x), \quad x \in \mathcal{X}$$

que satisfaz $p_X(x) \geq 0$ e $\displaystyle\sum_{x \in \mathcal{X}} p_X(x) = 1$.

#### **Variáveis Contínuas**

A variável $X$ é dita **contínua** se existir uma função $f_X : \mathbb{R} \to [0, \infty)$, chamada de **função densidade de probabilidade (PDF)**, tal que:

$$P(a \leq X \leq b) = \int_a^b f_X(x)\, dx$$

A PDF satisfaz $\displaystyle\int_{-\infty}^{\infty} f_X(x)\, dx = 1$. Note que $f_X(x)$ não é uma probabilidade, pois essa só é obtida integrando sobre um intervalo.

#### **Esperança**

Consiste no valor esperado de uma variável aleatória:

* **Caso Discreto**

Sendo $X$ uma variável discreta, então:

$$
E[X] = \sum_{x \in \mathcal{X}} x\, p_X(x)
$$

* **Caso Contínuo**

Sendo $X$ uma variável contínua, então:

$$
E[X] = \displaystyle\int_{-\infty}^{\infty} x \, f_{X}(x)dx
$$

#### **Variância**

Seja $\mu = E[X]$, a variância de $X$ mede a dispersão dos valores em torno da média:

$$
Var(X) = E[(X - \mu)^{2}]
$$

Uma forma computacional equivalente de determinar a variância é:

$$
Var(X) = E[X^{2}] - E[X]^{2}
$$

#### **Desvio Padrão**

Como $Var(X)$ está em unidades ao quadrado de $X$, define-se o
desvio padrão como:

$$
\sigma_{X} = \sqrt{Var(X)}
$$

que possui a mesma unidade de $X$, tornando-o mais interpretável como medida de dispersão.
