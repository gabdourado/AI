## **Redes Perceptron**

Dado um conjunto de observações

$$
\mathcal{D} = \{(\mathbf{x}_i, y_i)\}_{i=1}^N
\quad \text{onde}
\quad \mathbf{x}_i \in \mathbb{R}^{D} \text{ e } y_i \in \{-1, +1\}
$$

definimos o **vetor aumentado** $\tilde{\mathbf{x}}_i$ (com o bias absorvido):

$$
\tilde{\mathbf{x}}_i = \begin{bmatrix} 1 \\ \mathbf{x}_i \end{bmatrix}
$$

Seja então o vetor de pesos $\mathbf{w}$, definido por:

$$
\mathbf{w} =
\begin{bmatrix}
w_0 \\ w_1 \\ \vdots \\ w_D
\end{bmatrix}
$$

Um neurônio artificial recebe um vetor de entradas $\tilde{\mathbf{x}}$ e computa uma saída por meio de duas operações em sequência:

**Combinação linear (pré-ativação):**

$$
u = \mathbf{w}^{\top}\tilde{\mathbf{x}} = \sum_{i=0}^{D}w_i\tilde{x}_i
$$

**Aplicação da função de ativação:**

$$
y = \phi(u)
$$

Em redes perceptron, geralmente se utiliza a função degrau de Heaviside na forma bipolar:

$$
\phi(u) = \begin{cases}
+1, & \text{se } u \geq 0 \\
-1, & \text{se } u < 0
\end{cases}
$$

Durante o treinamento, ajustamos os pesos de acordo com a equação que segue:

$$
\mathbf{w}^{(t+1)} = \mathbf{w}^{(t)} + \eta\, \tilde{\mathbf{x}}_i\bigl(y_i - \hat{f}(\tilde{\mathbf{x}}_i)\bigr)
$$

onde $\eta > 0$ é a taxa de aprendizado, $y_i$ é o rótulo verdadeiro e $\hat{f}(\tilde{\mathbf{x}}_i) = \phi(\mathbf{w}^{(t)\top}\tilde{\mathbf{x}}_i)$ é a predição atual do modelo.
