## **Perceptron Multicamadas (MLP) — Notação Matricial**

No notebook anterior, desenvolvi o forward pass e o backpropagation em notação escalar, onde cada peso $w_{ji}^{(\ell)}$, ativação $z_j^{(\ell)}$ e delta $\delta_j^{(\ell)}$ eram tratados de maneira individual. Neste notebook, vou reescrever as equações em **notação matricial**, facilitando a implementação.

### **Pré-Ativação e Ativação**

No notebook anterior, **pré-ativação** e **ativação** de cada neurônio $j$ na camada $\ell$ eram dadas por:

$$
a_j^{(\ell)} = \sum_i w_{ji}^{(\ell)} \cdot z_i^{(\ell-1)}
\quad \text{e} \quad
z_j^{(\ell)} = h\left(a_j^{(\ell)}\right)
$$

Para reescrever isso em notação matricial, dada uma camada $\ell$ com $J$ neurônios recebendo $I$ entradas, definimos:

- O **vetor de ativações** $\mathbf{z}^{(\ell)} \in \mathbb{R}^{J}$, cujo $j$-ésimo elemento é $z_j^{(\ell)}$
- A **matriz de pesos** $\mathbf{W}^{(\ell)} \in \mathbb{R}^{J \times I}$, cujo elemento $(j, i)$ é $w_{ji}^{(\ell)}$

> A convenção $w_{ji}$ — "destino $j$ primeiro, origem $i$ depois" — garante que cada linha de $\mathbf{W}^{(\ell)}$ contém todos os pesos que chegam ao neurônio $j$.


Definimos ainda o **vetor de pré-ativações** $\mathbf{a}^{(\ell)} \in \mathbb{R}^{J}$ como a coleção de todos os $a_j^{(\ell)}$. Assim, o somatório escalar

$$
a_j^{(\ell)} = \sum_i w_{ji}^{(\ell)} \cdot z_i^{(\ell-1)}
$$

é o produto interno entre a $j$-ésima linha de $\mathbf{W}^{(\ell)}$ e o vetor $\mathbf{z}^{(\ell-1)}$. Empilhando todos os neurônios $j$ simultaneamente:

$$
\mathbf{a}^{(\ell)} = \mathbf{W}^{(\ell)}\, \mathbf{z}^{(\ell-1)}
$$

A função de ativação $h(\cdot)$ é aplicada elemento a elemento sobre $\mathbf{a}^{(\ell)}$:

$$
\mathbf{z}^{(\ell)} = h\!\left(\mathbf{a}^{(\ell)}\right)
$$

### **Forward Pass**

Aplicando as equações acima recursivamente para $\ell = 1, \ldots, L$, com $\mathbf{z}^{(0)} = \mathbf{x}$:

$$
\mathbf{a}^{(\ell)} = \mathbf{W}^{(\ell)}\mathbf{z}^{(\ell-1)}
\quad
\text{e}
\quad
\mathbf{z}^{(\ell)} = h\!\left(\mathbf{a}^{(\ell)}\right)
$$

A saída da rede é $\mathbf{y} = \mathbf{z}^{(L)}$.

#### **Dimensões**

Para não se perder, é útil verificar as dimensões em cada passo:

| Objeto | Shape |
|---|---|
| $\mathbf{W}^{(\ell)}$ | $J \times I$ |
| $\mathbf{z}^{(\ell-1)}$ | $I \times 1$ |
| $\mathbf{a}^{(\ell)}$ | $J \times 1$ |
| $\mathbf{z}^{(\ell)}$ | $J \times 1$ |## Da Notação Escalar para a Matricial

### **Backward Pass**

No notebook anterior, definimos o delta de cada neurônio $j$ na camada $\ell$ como:

$$
\delta_j^{(\ell)} \equiv \frac{\partial E_n}{\partial a_j^{(\ell)}}
$$

De forma análoga ao que fizemos no forward pass, definimos o **vetor de deltas** $\boldsymbol{\delta}^{(\ell)} \in \mathbb{R}^{J}$, cujo $j$-ésimo elemento é $\delta_j^{(\ell)}$.

Vamos pular o **passo 1** e seguir para o **passo 2**.

#### **Passo 2 — Delta na Camada de Saída**

Na camada de saída $L$, o delta escalar era:

$$
\delta_k^{(L)} = y_k - t_k
$$

Empilhando todos os $k$ neurônios de saída simultaneamente:

$$
\boldsymbol{\delta}^{(L)} = \mathbf{y} - \mathbf{t}
$$

onde $\mathbf{y} = \mathbf{z}^{(L)}$ é a saída da rede e $\mathbf{t}$ é o vetor de targets.

#### **Passo 3 — Retropropagação dos Deltas**

Para as camadas $\ell = (L-1), \ldots, 1$, o delta escalar era:

$$
\delta_j^{(\ell)} = h'\!\left(a_j^{(\ell)}\right) \sum_k w_{kj}^{(\ell+1)}\, \delta_k^{(\ell+1)}
$$

onde $K$ é o número de neurônios da camada $\ell+1$. O somatório
$\sum_k w_{kj}^{(\ell+1)} \delta_k^{(\ell+1)}$ é o produto interno entre a $j$-ésima
**coluna** de $\mathbf{W}^{(\ell+1)}$ e o vetor $\boldsymbol{\delta}^{(\ell+1)}$,
o que equivale a $\left(\mathbf{W}^{(\ell+1)}\right)^\top \boldsymbol{\delta}^{(\ell+1)}$.
Assim:

$$
\boldsymbol{\delta}^{(\ell)} = h'\!\left(\mathbf{a}^{(\ell)}\right) \odot \left(\mathbf{W}^{(\ell+1)}\right)^\top \boldsymbol{\delta}^{(\ell+1)}
$$

A transposição garante que os índices se alinham corretamente: $\mathbf{W}^{(\ell+1)} \in \mathbb{R}^{K \times J}$,
onde $K$ é o número de neurônios da camada $\ell+1$ e $J$ o da camada $\ell$. Então
$\left(\mathbf{W}^{(\ell+1)}\right)^\top \in \mathbb{R}^{J \times K}$,
resultando em $\boldsymbol{\delta}^{(\ell)} \in \mathbb{R}^{J}$.

#### **Passo 4 — Gradiente dos Pesos**

O gradiente escalar era:

$$
\frac{\partial E_n}{\partial w_{ji}^{(\ell)}} = \delta_j^{(\ell)} \cdot z_i^{(\ell-1)}
$$

Esse produto entre um escalar $\delta_j^{(\ell)}$ e um escalar $z_i^{(\ell-1)}$ gera o elemento $(j, i)$ da matriz gradiente. Empilhando todos os pares $(j, i)$ simultaneamente, isso é reconhecido como um produto externo:

$$
\frac{\partial E_n}{\partial \mathbf{W}^{(\ell)}} = \boldsymbol{\delta}^{(\ell)} \otimes \left(\mathbf{z}^{(\ell-1)}\right)^\top
$$

onde $\frac{\partial E_n}{\partial \mathbf{W}^{(\ell)}} \in \mathbb{R}^{J \times I}$ tem o mesmo shape que $\mathbf{W}^{(\ell)}$ — o que é necessário para a atualização fazer sentido.

#### **Passo 5 — Atualização dos Pesos**

Com o gradiente matricial em mãos, a atualização via SGD puro fica:

$$
\mathbf{W}^{(\ell)} \leftarrow \mathbf{W}^{(\ell)} - \eta\, \boldsymbol{\delta}^{(\ell)} \left(\mathbf{z}^{(\ell-1)}\right)^\top
$$
