## **Busca Semântica (Information Retrieval)**

Dado o dataset [**20 Newsgroups**](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html) e uma query, vamos encontrar os 3 documentos mais semelhantes.

Para isso, vamos representar tanto os documentos como a query como vetores TF-IDF no mesmo espaço vetorial. Em seguida, calcular a similaridade cosseno entre a query e cada documento do dataset. Por fim, vamos ranquear os textos e retornar os mais relevantes, de forma semelhante a um motor de busca.


```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import numpy as np
```

```python
newsgroups = fetch_20newsgroups(
    subset="all",
)

documents = newsgroups.data
labels = newsgroups.target

print(f"Number of docments: {len(documents)}\n")

print("The groups:")
for group in newsgroups.target_names:
  print(group)
```

```python
vectorizer = TfidfVectorizer(stop_words="english")

X = vectorizer.fit_transform(documents)
```

```python
print(f"The Document-Term Matrix is: {X.shape[0]} x {X.shape[1]}")
```

```python
query = ["I'm atheist."]

q_vec = vectorizer.transform(query)
```

```python
similarities = cosine_similarity(q_vec, X)

top_k = 3

indices = np.argsort(similarities[0])[-top_k:][::-1]

for i in indices:
    print(f"Score: {similarities[0][i]}")
    print(documents[i][:500])
    print("--------------------------\n")
```
